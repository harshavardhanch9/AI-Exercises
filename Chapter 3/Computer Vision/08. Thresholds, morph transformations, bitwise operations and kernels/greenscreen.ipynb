# Green Screen mini project

```python
import cv2
import matplotlib.pyplot as plt
```

```python
bgr_img = cv2.imread('img/green screen.png')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))
plt.imshow(rgb_img)
```

```python
hsv_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2HSV)
lower_bounds = (40, 0, 30)
upper_bounds = (80, 255, 255)

mask = cv2.inRange(hsv_img, lower_bounds, upper_bounds)

plt.figure(figsize=(15,15))
plt.imshow(mask, cmap = 'gray')
```

```python
img = rgb_img.copy()
img[mask==255]=[0, 0, 0]
plt.figure(figsize=(15,15))
plt.imshow(img, cmap = 'gray')
```

```python
back_ground = cv2.imread('img/background.png')
plt.figure(figsize=(15,15))
plt.imshow(back_ground, cmap = 'gray')
```

```python
background = back_ground[0:370]
background[mask==0] = [0,0,0]
plt.figure(figsize=(15,15))
plt.imshow(background, cmap = 'gray')
```

```python
both_img = img + background
rgb_final_img = cv2.cvtColor(both_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))
plt.imshow(rgb_final_img)
```
