# QR Code Detector

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/qr code.png')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(15,15))
plt.imshow(rgb_img)
```

```python
CodeDetector = cv2.QRCodeDetector()
text, points, code = CodeDetector.detectAndDecode(img)

if points is not None:
    print('QR detected successfully')
    print(f'Text: {text}')
    
```

```python
img_copy = rgb_img.copy()
cv2.rectangle(img_copy, (34,78), (140,175), (0,225,0),2)
plt.figure(figsize=(20,15))
plt.imshow(img_copy)
```

```python
cv2.imwrite('img/qr1.png', code)
```

```python

qr_code = cv2.imread('img/qr1.png')
rgb_qr = cv2.cvtColor(qr_code, cv2.COLOR_BGR2RGB)
hsv_qr = cv2.cvtColor(qr_code, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(10,10))
plt.imshow(rgb_qr)
```
