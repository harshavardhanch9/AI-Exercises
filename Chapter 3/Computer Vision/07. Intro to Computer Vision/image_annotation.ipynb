# Image Annotation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/bike.jpg')

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,15))
plt.imshow(rgb_img)
```

```python
img_copy = rgb_img.copy()
cv2.rectangle(img_copy, (80,80),(235,200),(0,255,0),2)
cv2.putText(img_copy, 'Car', (80,75), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),2)

cv2.rectangle(img_copy, (290,40),(430,220),(255,255,255),2)
cv2.putText(img_copy, 'Bike', (290,35), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255),2)

plt.figure(figsize=(15,15))
plt.imshow(img_copy)
```
